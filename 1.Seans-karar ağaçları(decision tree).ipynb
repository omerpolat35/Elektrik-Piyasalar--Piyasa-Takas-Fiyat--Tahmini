### Eğitim / Test verisine bölme

```python
from sklearn.model_selection import train_test_split
""" Not: x değişkeni herkes tarafından bilinen demand, yakıt fiyatları vb. olmalıdır. Fiyat tahmini yapılacak seneryoda 
oyuncuların verdidiği teklikler bilinenler içerisinde olmadığında x parametresine atanmamalıdır.
x bağımsız değişkenleri tek tek girilmelidir. Burada daha önce oluşturulan kolerasyon ve descipe çıktıları incelenerek 
karar verilebilir. Sonuca etkisi olmadığı düşünülen kısımların eklenmemesi tercih edilmelidir.
y = wx + b
y -> label """
y = mPCorr["marketPrice"].values
# x -> feature (özellik)
x = mPCorr[["oldMarketPrace","player kazanma","demand","player_bid_price","fc_Urn","period","cc_Coal","fc_Coal","investmentCost","Bioenergy"]].values
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=15)
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtReg = DecisionTreeRegressor(max_depth=3, random_state=0)
#dtReg1 = DecisionTreeClassifier(criterion ='entropy',random_state=0)
dtReg.fit(x_train, y_train)
y_tahmin = dtReg.predict(x_test)
```

### R^2 / R-squared - Modelleri karşılaştırırken, daha yüksek R kare değerine sahip model, veriler için daha uygundur.
Belirleme katsayısı olarak da bilinen R kare, verilerin uyan regresyon çizgisine ne kadar yakın olduğunu gösteren bir ölçüdür.

Bağımsız değişkenler den tahmin edilebilen bağımlı değişkendeki varyansın oranıdır.

R-karenin değeri, doğrusal bir modelle açıklanan yanıt değişkeninin (y) değişim yüzdesidir.

(ytest (i) - preds (i)) ** 2 . y gözlenen her değerdir y (i) eksi gözlemlenen değerlerin ortalamasıdır np.ortalama (ytest)
### Mean Squared Error (MSE) - Modelleri karşılaştırırken, en küçük MSE değerine sahip model veriler için daha uygundur.
Ortalama Kare Hata, hataların karelerinin ortalamasını, yani gerçek değer (y) ile tahmini değer (ŷ) arasındaki farkı ölçer.

Bu durumda hata , gözlemlenen değerler y1, y2, y3,.... ile tahmin edilen değerler arasındaki fark anlamına gelir pred (y1), pred 
(y2), pred (y3),....

Her farkı kareleriz (pred (yn) - yn) ) ** 2, böylece negatif ve pozitif değerler birbirini götürmez.

```python
print('Mean Squared Error_dtReg1:', metrics.mean_squared_error(y_test, y_tahmin)) 
print('RSquared_dtReg1:', r2_score(y_test, y_tahmin))
```

Decision Tree, daha düşük bir MSE ve daha yüksek bir R-kare ile sonuçlandığından tahmin etmek için GradientBoosting daha uygun bir model olduğu sonucuna varabiliriz.
