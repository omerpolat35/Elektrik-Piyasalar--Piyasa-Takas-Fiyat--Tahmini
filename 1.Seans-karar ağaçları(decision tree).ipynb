### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtReg = DecisionTreeRegressor(max_depth=3, random_state=0)
#dtReg1 = DecisionTreeClassifier(criterion ='entropy',random_state=0)
dtReg.fit(x_train, y_train)
y_tahmin = dtReg.predict(x_test)
```

### R^2 / R-squared - Modelleri karÅŸÄ±laÅŸtÄ±rÄ±rken, daha yÃ¼ksek R kare deÄŸerine sahip model, veriler iÃ§in daha uygundur.
Belirleme katsayÄ±sÄ± olarak da bilinen R kare, verilerin uyan regresyon Ã§izgisine ne kadar yakÄ±n olduÄŸunu gÃ¶steren bir Ã¶lÃ§Ã¼dÃ¼r.

BaÄŸÄ±msÄ±z deÄŸiÅŸkenler den tahmin edilebilen baÄŸÄ±mlÄ± deÄŸiÅŸkendeki varyansÄ±n oranÄ±dÄ±r.

R-karenin deÄŸeri, doÄŸrusal bir modelle aÃ§Ä±klanan yanÄ±t deÄŸiÅŸkeninin (y) deÄŸiÅŸim yÃ¼zdesidir.

(ytest (i) - preds (i)) ** 2 . y gÃ¶zlenen her deÄŸerdir y (i) eksi gÃ¶zlemlenen deÄŸerlerin ortalamasÄ±dÄ±r np.ortalama (ytest)
### Mean Squared Error (MSE) - Modelleri karÅŸÄ±laÅŸtÄ±rÄ±rken, en kÃ¼Ã§Ã¼k MSE deÄŸerine sahip model veriler iÃ§in daha uygundur.
Ortalama Kare Hata, hatalarÄ±n karelerinin ortalamasÄ±nÄ±, yani gerÃ§ek deÄŸer (y) ile tahmini deÄŸer (Å·) arasÄ±ndaki farkÄ± Ã¶lÃ§er.

Bu durumda hata , gÃ¶zlemlenen deÄŸerler y1, y2, y3,.... ile tahmin edilen deÄŸerler arasÄ±ndaki fark anlamÄ±na gelir pred (y1), pred 
(y2), pred (y3),....

Her farkÄ± kareleriz (pred (yn) - yn) ) ** 2, bÃ¶ylece negatif ve pozitif deÄŸerler birbirini gÃ¶tÃ¼rmez.
